# Pandas - Merging , Joining and Concatenate

```python
import pandas as pd
```

```python
df1 = pd.read_csv(r"C:\Users\battu\Documents\Python Jupyter notebook Files\Raw Data\LOTR.csv")
```

```python
df2 = pd.read_csv(r"C:\Users\battu\Documents\Python Jupyter notebook Files\Raw Data\LOTR 2.csv")
```

```python
df1
```

```python
df2
```

```python
df1.merge(df2) # this is going to do an inner join by default
```

```python
df1.merge(df2, how = 'inner', on = 'FellowshipID')
```

```python
df1.merge(df2, how = 'inner', on = ['FellowshipID','FirstName'])
```

```python
df1.merge(df2, how = 'outer')
```

```python
df1.merge(df2, how = 'left')
```

```python
df1.merge(df2, how = 'right')
```

```python
df1.merge(df2, how = 'cross')
```

```python
df1.join(df2, on = 'FellowshipID', how = 'outer', lsuffix='_Left', rsuffix='_Right')
```

```python
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'),lsuffix='_Left', rsuffix='_Right', how = 'outer')
df4
```

```python
#Concatenate
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2], join = 'inner')
```

```python
pd.concat([df1,df2], join = 'outer')
```

```python
pd.concat([df1,df2], join = 'outer', axis = 1)
```
