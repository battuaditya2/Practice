# Scrapping data from websites 1

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_social_platforms_with_at_least_100_million_active_users'
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
# <table class="wikitable sortable jquery-tablesorter">
```

```python
soup.find_all('table')
```

```python
table = soup.find('table', class_ = 'wikitable sortable')
```

```python
print(table)
```

```python
main_titles = table.find_all('th')
```

```python
print(main_titles)
```

```python
table_titles = [titles.text.strip() for titles in main_titles]
print(table_titles)
```

```python
import pandas as pd
df = pd.DataFrame(columns = table_titles)
df
```

```python
column_data = table.find_all('tr')
```

```python
column_data
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    length = len(df)
    df.loc[length] = individual_row_data
```
